```python
# prompt: i want to extract id from image name, the id is basically the name, i will give the images folder and it will put it in a list

folder = "/content/drive/MyDrive/IndabaX_2024_Files/train"
import os
def extract_ids(images_folder):
  ids = []
  for filename in os.listdir(images_folder):
    id = filename.split('.')[0]
    ids.append(id)
    ids.sort()
  return ids
```

```python
images = extract_ids(folder)
print(images)
```

```python
# prompt: import regex and use it to get the image with her id

import re

# Define the regular expression to match the image ID
pattern = r"ID_(\d+)"

# Initialize the image path variable
image_path = None
ids = []
# Iterate through the list of image IDs
for image_id in images:
  # Check if the image ID matches the regular expression
  sub = re.findall(pattern, image_id)
  ids.append(int(sub[0]))

print(ids)
```
